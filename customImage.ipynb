```python
import numpy as np
import cv2
```

```python
custom = np.zeros((300,300,3),dtype = "uint8")
```

```python
custom = cv2.rectangle(custom,(50,50),(250,250),(255,0,0),10)
custom = cv2.rectangle(custom,(100,100),(200,200),(0,255,0),10)
custom = cv2.rectangle(custom,(150,150),(150,150),(0,0,255),10)
```

```python
#custom[20:280,20:220] = [[243]]
cv2.imshow("image",custom)
cv2.waitKey()
cv2.destroyAllWindows()
```
